```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\MY PC\Downloads\Employee attrition.csv")

plt.style.use("ggplot")
```

```python
print("1. Attrition Count")
print(df['Attrition'].value_counts(), "\n")

print("2. Gender-wise Attrition")
print(df.groupby('Gender')['Attrition'].value_counts(), "\n")

print("3. Age-wise Attrition (Top 10 Ages)")
print(df.groupby('Age')['Attrition'].value_counts().head(10), "\n")

print("4. Department-wise Attrition")
print(df.groupby('Department')['Attrition'].value_counts(), "\n")

print("5. Job Role-wise Attrition")
print(df.groupby('JobRole')['Attrition'].value_counts(), "\n")

print("6. Overtime vs Attrition")
print(df.groupby('OverTime')['Attrition'].value_counts(), "\n")

print("7. Monthly Income: Attrition vs Non-Attrition")
print(df.groupby('Attrition')['MonthlyIncome'].mean(), "\n")
```

```python
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title("Attrition by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', hue='Attrition', kde=True)
plt.title("Attrition by Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='JobRole', hue='Attrition', data=df)
plt.title("Attrition by Job Role")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.show()
```

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Attrition vs Overtime")
plt.xlabel("OverTime")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.show()
```
